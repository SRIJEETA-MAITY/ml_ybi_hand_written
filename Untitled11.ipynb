```python
labels=[]
file_names=os.listdir('/content/imageresized')
for i in range(10):
  file_name=file_names[i]
  if file_name[0:3]=='pik':
    labels.append(1)
  else:
    labels.append(0)
print(labels)
print(file_names)
```

```python
import tensorflow as tf
import tensorflow_hub as hub
```

```python
mobilenet_model='https://tfhub.dev/google/tf2-preview/mobilenet_v2/feature_vector/4'
```

```python
pretrained_model=hub.KerasLayer(mobilenet_model,input_shape=(224,224,3),trainable=False)
```

```python
num_class=2
model=tf.keras.Sequential()
model.add(tf.keras.layers.Lambda(lambda x: pretrained_model(x)))

model.add(tf.keras.layers.Dense(num_class))
model.summary()
model.compile(
    optimizer='adam',
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics=['acc']
)
```

```python
model.fit(x,y,epochs=5)
```

```python
input_image_path=input('path of image to be predicted:')
input_image=cv2.imread(input_image_path)
cv2_imshow(input_image)
input_image_resize=cv2.resize(input_image,(224,224))
input_scaled=input_image_resize/255
input_reshaped=np.reshape(input_scaled,[1,224,224,3])
input_prediction=model.predict(input_reshaped)
input_pred_label=np.argmax(input_prediction)
if input_pred_label==0:
  print('The image is raichu')
else:
  print('The image is pikachu')
```
